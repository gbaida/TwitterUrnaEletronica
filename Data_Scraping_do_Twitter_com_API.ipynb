#Importando tweets em um DataFrama

```python
#Query de busca https://developer.twitter.com/en/docs/twitter-api/enterprise/rules-and-filtering/operators-by-product
query = 'Urna eletrônica -is:retweet '

#lista temporária
lista = []
#Rodando o método da API
for tweet in tweepy.Paginator(client.search_recent_tweets, query=query,
                              tweet_fields=['author_id','created_at', 'geo', 'lang', 'public_metrics'], max_results=100).flatten(limit=10000):
  #Adicionando tweets na lista
  lista.append([tweet.author_id,tweet.text, tweet.created_at, tweet.geo, tweet.lang, tweet.public_metrics])

#Criando o dataframe
tweets_df = pd.DataFrame(lista, columns=['author_id','text','created_at', 'geo', 'lang', 'public_metrics'])
tweets_df
```

```python
#Quebrando o campo Public Metrics em Colunas diferentes
tweets_df = pd.concat([tweets_df.drop(["public_metrics"], axis=1), pd.json_normalize(tweets_df['public_metrics'])], axis=1)
tweets_df
```

```python
#Importando o Modelo Roberta
from google.colab import output
output.enable_custom_widget_manager()
#Importanto bibliotecas Necessárias para o Modelo
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from scipy.special import softmax
```

Modelo de Processamento de Linguagem Natural RoBERTa
https://ai.facebook.com/blog/roberta-an-optimized-method-for-pretraining-self-supervised-nlp-systems/

```python
roberta = "cardiffnlp/twitter-roberta-base-sentiment"
model = AutoModelForSequenceClassification.from_pretrained(roberta)
tokenizer = AutoTokenizer.from_pretrained(roberta)
```

##Rodando modelo no DataSet

```python
from collections import defaultdict
```

```python

tweet_list = []
#model_dict = {}
model_list = []
#Pré Processamento dos tweets para serem lidos pelo modelo
model_dict = defaultdict(list)
for tweet in tweets_df['text']:
  tweet_words = []
  for word in tweet.split(' '):
      if word.startswith('@') and len(word) > 1:
          word = '@user'

      elif word.startswith('http'):
          word = "http"
      tweet_words.append(word)
  tweet_proc = " ".join(tweet_words)

  #Codificando o tweet pra ser lido pelo modelo
  encoded_tweet = tokenizer(tweet_proc, return_tensors='pt')
  output = model(**encoded_tweet)

  #Quebrando o Output do Modelo em Scores
  labels = ['Negative', 'Neutral', 'Positive']
  scores = output[0][0].detach().numpy()
  scores = softmax(scores)
  #Loop Adicionando os Scores ao Dicionário
  for i in range(len(scores)):
        l = labels[i]
        s = scores[i]
        model_dict[l].append(s)
model_dict
```

Adicionando os Valores Previstos como Novas Colunas ao DF Original

```python
#Transformando o dicionário em um Dataframe
df_dict = pd.DataFrame(model_dict)
#Concatenando os Dataframes
tweets_df = pd.concat([tweets_df, df_dict], axis=1)
tweets_df
```
